## Understanding pointers

You can think about a pointer as an address.

For instance, you can save in your organizer the addresses of your friends.

Your orgnizer does not contain your friends. It only cointains a list of addresses and phone numbers where you find your friends.

Likewise, when you create a variable in Python, a space in memory is allocated to store your data at a specific address.

For instance, when you write:

```
a = 10
```

A space in your computer memory will be allocated to store the value (10) or your variable (a).

```
b = a
```

When you define a second variable (b) and allocate the value a to this new variable, you are telling Python not to reserve another space in memory but instead, to re-use the space already allocated to variable a.

Lastly, if you decide to re-use variable a to store a different piece of data:

```
a = "John"
```

then Python will allocate a new space in memory to store the value "John" for variable a. Variable a will use a new address pointing to the space in memory where "John" is stored.

Meanwhile, variable b is still pointing to the same address and therefor, the value associated with variable b has not changed.

```
print(b)
```

returns 10.



<img src="https://nico.nexgate.ch/images/pointers.png" width='80%' alt="excel" />

````python
# Code example to illustrate the concept of pointers in Python

a = 10
print("a = ",a, " -> Python is allocating via a pointer a space in memory where the value 10 is stored")
b = a
print("b = ",b, " -> because b points to the same place in memory as a")
a = "John"
print("a = ",a," -> now a has a new value and this new value is stored in a new space in memory")
print("b = ",b," -> because b still points to its original space in memory, therefor the value of b has not changed")
````

#### Resolving quadratic equations

$a x^2+b x+c=0$

First, calculate $\Delta = b^2-4ac$

If $\Delta < 0$ return "no solution"$

If $\Delta > 0$ return 2 solutions $x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}$

Otherwise, if $\Delta = 0$ then the solution is $x=\frac{-b}{2a}$

````python
import math


a, b, c = 1, 2, -15

delta = (b*b) - (4 * a * c)
if delta < 0:
    print('no solution')
elif delta == 0:
    print ('x =', (-b) / (2 * a))
else: # useless else
    print ('x1 =', ((-b) + math.sqrt(delta)) / (2 * a), 'x2 =', ((-b) - math.sqrt(delta)) / (2 * a))
````

üå∂Ô∏è <font color='red'>TODO</font>: Write a function resolve_equation(a,b,c) that resolves the quadratic equation $a x^2+b x+c=0$

````python
import math
from typing import Optional

def resolve_equation(a: float, b: float, c: float) -> Optional[tuple[float]]:
    """
    Determines the two solutions for the quadratic equations a*x^2 + b*x + c = 0

    Arguments:
        a (float): Value for variable a in the equation
        b (float): Value for variable b in the equation
        c (float): Value for variable c in the equation

    Returns:
        None:           If no solution exists, None is returned.
        (float, ):      The only solution for the quadratic equations (as tuple with 1 element).
        (float, float): The two solutions for the quadratic equations.
    """
    delta = (b*b) - (4 * a * c)
    if delta < 0:
        print('delta is smaller than 0')
        return None
    elif delta == 0:
        return ((-b) / (2 * a),)
    else: # useless else
        return (((-b) + math.sqrt(delta)) / (2 * a), ((-b) - math.sqrt(delta)) / (2 * a))
````

## üå∂Ô∏è Quadratic equation revisited

The code below shows you how to use an existing funciton from the Numpy package to solve a quadratic equation.

test the code forthis equation: $2 x^2+10 x-12=0$

(a = 2, b = 10 and c = -12)

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
a = float(input('Enter a: '))  
b = float(input('Enter b: '))  
c = float(input('Enter c: '))  

np.roots([a,b,c])
````

````python
def f(a,b,c,x):
    return a * x * x + b * x + c


f(a,b,c,2)
````

From the Numpy package, use the linspace() function to create a list of numbers from a to b. 

points.png

For instance, you can create 21 values between -10 and 10 using the code below:

````python
# 50 linearly spaced numbers
x = np.linspace(-10,10,21)
x
````

You can then create plots for a chart by calculating the value of f(x) for all points defined in your range.

````python
y = f(a,b,c,x)

print("The values of f(x) are: ",y)
````

And finally, you can use the Matplotlib package to plot your equation.

NOTE: we will cover visualization in more details in week 4. 

````python
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.ylim(-100, 100)
plt.xlim(-7,7)

plt.xlabel("x")
ax.xaxis.set_label_coords(0.998,0.57)

plt.text(4.5, f(a,b,c,4.5), '    $2 x^2+10 x-12=0$')

# plot the quadratic function
plt.plot(x,y, 'b:')

plt.title("Quadratic equation", loc = 'left')

plt.show();
````
