## Explore the data and examples of special types of variables
nominal variables
* Street
* Sale Type

ordinal variables
* Order
* Year Sold

discrete variables
* Mo Sold

continuous variables
* Lot Frontage?

```python
# ChatGPT's solution:
# nominal variables
nominal_cols = [col for col in df.columns if df[col].nunique() < 10]
print('Nominal variables: ', nominal_cols)

# ordinal variables
ordinal_cols = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique() > 10]
print('Ordinal variables: ', ordinal_cols)

# discrete variables
discrete_cols = [col for col in df.columns if df[col].dtype == 'int64']
print('Discrete variables: ', discrete_cols)

# continuous variables
continuous_cols = [col for col in df.columns if df[col].dtype == 'float64']
print('Continuous variables: ', continuous_cols)
```
