üå∂Ô∏è TODO 2:

- write some code that uses your variables to output:
    
<blockquote>Hi John, you are 36 years old</blockquote>

```python
import datetime
from dateutil.relativedelta import relativedelta

today_as_dt = datetime.datetime.strptime(today, '%d-%m-%Y') # would be better: datetime.datetime.now()
date_of_birth_as_dt = datetime.datetime.strptime(date_of_birth, '%d-%m-%Y')

age_in_years = relativedelta(today_as_dt, date_of_birth_as_dt).years

print(f"Hi {name}, you are {age_in_years} years old")
```

üå∂Ô∏è TODO 3: 
- Concatenate these 2 arrays and assign the result to a variable z
- Then calculate the mean (average) and median values
- Sort the array
- Add comments to your code


```python
import numpy as np

# Given: the two arrays:
x = np.array([10, 17, 13, 21, 14, 15])
y = np.array([12, 26, 10])

# Put the values of x and y in only 1 array:
z = np.concatenate([x, y])

# Calculate the mean and median value of the array z:
mean = z.mean()
median = np.median(z)

print('mean:', round(mean, 1), 'median:', median)

# And now sort the whole array:
z.sort()

print('sorted:', z)
```
