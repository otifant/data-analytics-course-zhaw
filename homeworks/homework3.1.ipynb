# Partial homework solution

2430 observations with potentially 80 features + 2 identifiers (Order, PID)
The identifers do not have any influence on the dependent variable (SalePrice) and therefor will be removed form the training data set.

There is a mix of continuous and categorical varibles which can be further split into 4 groups:
- 23 nominal variables (e.g. 'MS Zoning' qualifies the purpose such as commercial vs. residential)
- 23 ordinal variables (e.g. 'Garage Cond' from 'no garage', 'poor', ... to 'excellent')
- 14 discrete variables (e.g. 'Fireplaces' counts the number of fireplaces)
- 20 continuous variables (e.g. 'Gr Liv Area' which measures the living area in square feet)

````python
import pandas as pd

df = pd.read_csv('https://bit.ly/3uiJFuH')
````

````python
df.head()
````

````python
# removing the 2 columns containing identifiers
print("shape before removing the columns ORder and PID ",df.shape) # (2430,82)
df.drop(['Order','PID'],axis=1,inplace=True) # step 1 in preprocessing
print("shape after removing the columns ORder and PID ",df.shape) # (2430,80)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(df['Gr Liv Area'],df.SalePrice,label='normal')
liv_area_above_4000 = df['Gr Liv Area']>4000
plt.scatter(df[liv_area_above_4000]['Gr Liv Area'],
            df[liv_area_above_4000].SalePrice,
            c='C1',label='outliers')
plt.legend(loc='upper left')
plt.xlabel("Gr Liv Area")
plt.ylabel("Sale Price")
plt.show();
````

````python
import numpy as np
from scipy import stats
outliers_mask = np.abs(stats.zscore(df['Gr Liv Area'])) <= 3   
````
