### Creating a mask (filter) to remove outliers.

- step 1: calculate the z-scope for all values in the "Gr Liv Area" column
- step 2: define a mark variable which will contain a Series of True/False values for each row of data by checking if the z-score if between -3 and +3. In order to do that, we can get the absolute value of the z-score (i.e. the z-score with the + or - sign), then compare it to <= 3.
- step 3: apply this boolean condition to the dataframe 

````python
import numpy as np
from scipy import stats
````

````python
# Step 1
z_scores = stats.zscore(df['Gr Liv Area'])
z_scores
````

````python
# Step 2
outliers_mask = np.abs(stats.zscore(df['Gr Liv Area'])) <= 3   
outliers_mask
````

````python
# Step 3, use the outliers' mark to filter data
````

````python
df[outliers_mask]
````

## üå∂Ô∏è How to calculate the z-score from scratch

- step 1: calculate the standard deviation
- step 2: calcualte the z-score

````python
# Calculate the Standard Deviation 

# define a variable to hold the Pandas Series with all values for "Gr Liv Area"
Gr_Liv_Area_series = df['Gr Liv Area']

# calculate the mean average

mean_Gr_Liv_Area = sum(Gr_Liv_Area_series) / len(Gr_Liv_Area_series)

# calculate the differences and the sum of differences

Gr_Liv_Area_differences = [(value - mean_Gr_Liv_Area)**2 for value in Gr_Liv_Area_series]

sum_of_differences = sum(Gr_Liv_Area_differences)

std_Gr_Liv_Area = (sum_of_differences / (len(Gr_Liv_Area_series) - 1)) ** 0.5

print("The standard deviation for Gr Liv Area is ",std_Gr_Liv_Area)
````

````python
# we can verify that the std deviation calculation is correct by using the .describe() function on the dataframe 
df_cont.describe()[['Gr Liv Area']]
````

````python
# finally, we can calculate the z-score
zscores_Gr_Liv_Area = [(x - mean_Gr_Liv_Area) / std_Gr_Liv_Area for x in Gr_Liv_Area_series]

print(zscores_Gr_Liv_Area)
````

## üå∂Ô∏è Feature engineering

Feature engineering is a technique commonly used to prepare a dataset for machine learning. Let's look at two examples of feature engineering:

- creating binary (0 or 1) variables from a zero-inflated variable
- creating binary variables from qualitative variables

### üå∂Ô∏è Dealing with zero-inflated variables

For some variables, there is a large number of rows with the value 0. These are commonly refered to as zero-inflated variables. Let's look at an example: the "Pool Area" column contains many rows = 0. Let's visualize the distribution of this variable with and without the zero values.

````python
# with the zero values

plt.figure(figsize = (10,6))

sns.histplot(data = df, x = "Pool Area")

plt.show()
````

````python
# let's create a mask to filter our the zero values
filter_Pool_Area = df['Pool Area']>0
````

````python
# let's show the distribution without the zero values

plt.figure(figsize = (10,6))

sns.histplot(data = df[filter_Pool_Area], x = "Pool Area")

plt.show()
````

In this situation, we can create a new column to indicate if the property has a pool or not. This is called feature engineering because we create a new piece of data from existing data.

````python
df['has_Pool_Area']=np.where(filter_Pool_Area,1,0)
````

````python
df['has_Pool_Area']
````

````python
df['has_Pool_Area'].value_counts()
````
